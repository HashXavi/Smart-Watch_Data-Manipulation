### CREATING A COLUMN TO MEASURE THE STRESS LEVEL 

```python
## using numpy 

conditions = [
    (df['Stress Level']>=1) & (df['Stress Level'] <6),
    (df['Stress Level']>=6) & (df['Stress Level'] <8)

]

choices = ['Excellent','Satisfactory']

df['Stress_Ratings'] = np.select(conditions,choices, default='Bad')
```

#### CREATING COLUMN FOR OVERALL USER RATINGS

```python
bo_healthy = (df['Blood Oxygen Level (%)']>=94)&(df['Blood Oxygen Level (%)']<100)
bp_healthy = (df['Heart Rate (BPM)']>=50)&(df['Heart Rate (BPM)']<80)
sc_healthy = (df['Step Count']>=4000)&(df['Step Count']<15000)
sleep_health = (df['Sleep Duration (hours)']>=6)&(df['Sleep Duration (hours)']<8)

healty_count = bo_healthy.astype(int) + bp_healthy.astype(int) + sc_healthy.astype(int)+sleep_health.astype(int)

df['User_Ratings'] = np.select(
[healty_count == 4 , healty_count == 3,healty_count == 2],
['Healthy','Mid-Healthy','Boderline'],
default= 'Unhealthy'
)
```
